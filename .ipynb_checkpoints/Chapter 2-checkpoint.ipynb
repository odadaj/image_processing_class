### 2.1 Digital Image
A digital image comprises M x N pixels. Each pixel, represented by k bits, can exhibit 2k shades of gray in grayscale images. Pixel values range from 0 (black) to 2k-1 (white). An image's quality and resolution are determined by its pixel count.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import acos, sqrt, pi
img = cv2.imread("baboon.png")


# 1. Digital Image Representation
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title(f"Resolution: {gray_img.shape[0]} x {gray_img.shape[1]}")
plt.show()
```

### 2.2 Sampling and Quantization
When an image is sensed and converted into digital form, the image is discretized in two ways: sampling and quantization. Sampling converts coordinate values, while Quantization digitizes amplitude values. Image quality hinges on these processes. While image resolution concerns the number of samples, graylevel resolution pertains to the diversity of gray shades. Image storage space is influenced by these factors.

The number of samples taken in digitization is known as image resolution.
Quantization decides the accuracy of the pixel intensities.

### 2.3 Color Models
Visible light, a spectrum from 400 to 700 nm, comprises a range of colors. Newton's prism experiment illustrated this spectrum, featuring the VIBGYOR colors. Objects reflect light of specific wavelengths, determining their perceived color.

#### 2.3.1 RGB Color Model
The RGB model is additive, mixing light spectrums. Screens, for example, mix red, green, and blue phosphors to generate colors. This model is represented as a color cube with black at (0,0,0) and white at (1,1,1).



```python
colored_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(colored_img)
plt.title(f"Resolution: {colored_img.shape[0]} x {colored_img.shape[1]}")
plt.show()
```

#### 2.3.2 CMY Color Model
Cyan, magenta and yellow are the secondary colors of light and the primary colors of pigments. This means, if white light is shined on a surface coated with cyan pigment, no red light is reflected from it. Cyan subtracts red light from white light. Unlike the RGB color model, CMY is subtractive, meaning higher values are associated with darker colors rather than lighter ones.

The subtractive CMY model involves light reflection rather than emission. In printing, colors are combined to approach black. This model adds a K (black) value, leading to the CMYK model.

```python
# RGB to CMY
def rgb_to_cmy(rgb):
    cmy = 1 - rgb/255.0
    return cmy * 255

cmy_img = rgb_to_cmy(colored_img) 
plt.imshow(cmy_img/255.0)
plt.title("CMY Color Model")
plt.show()
```

#### 2.3.3 YIQ Color Model
YIQ caters to human visual sensitivity, prioritizing luminance over hue or saturation. It's analogous to YUV but with different mathematical representations.

#### 2.3.4 HSI Color Model
HSI encompasses hue, saturation, and intensity. The model visualizes all colors within a triangle formed by red, green, and blue vertices. The RGB-HSI conversion formula has been provided.

```python
# 3. Color Models

# RGB to HSI
def rgb_to_hsi(rgb):
    r, g, b = rgb[:, :, 0], rgb[:, :, 1], rgb[:, :, 2]
    r /= 255
    g /= 255
    b /= 255
    num = 0.5 * ((r - g) + (r - b))
    den = np.sqrt((r - g)**2 + (r - b) * (g - b))
    h = np.arccos(num / (den + 1e-6))
    h[b > g] = 2*pi - h[b > g]
    h /= 2*pi
    s = 1 - 3 * np.min([r, g, b], axis=0)
    i = (r + g + b) / 3
    return h, s, i


# h, s, i = rgb_to_hsi(colored_img)


# plt.imshow(h, cmap='hsv')
# plt.title("HSI - Hue Component")
# plt.show()
```

### 2.4 Basic Relationships between Pixels
Pixels have various relationships based on their neighboring pixels. 4-neighbors include vertical and horizontal neighbors, while 8-neighbors combine 4-neighbors and diagonal ones. The representation is different for boundary pixels.


```python
def show_neighbors(x, y, gray_img):
    plt.imshow(gray_img, cmap='gray')
    plt.scatter([y], [x], c='r') # Current pixel
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if 0 <= x+dx < gray_img.shape[0] and 0 <= y+dy < gray_img.shape[1] and (dx, dy) != (0, 0):
                plt.scatter([y+dy], [x+dx], c='b')  # Neighbor pixels
    plt.title("8-neighbors of the pixel")
    plt.show()

show_neighbors(350, 50, gray_img)
```

### 2.5 Adjacency
Adjacency defines the relationship between two pixels based on their neighboring status. Two pixels can be 4- or 8-adjacent, depending on their proximity and matching values in binary images.

##### 4-Adjacency:
4-Adjacent: Two pixels are 4-adjacent if they are 4-neighbors and have the same value in a binary image.
Simple Example: In a binary image, if a white pixel is right next to another white pixel horizontally or vertically, they are 4-adjacent.

##### 8-Adjacency:
8-Adjacent: Two pixels are 8-adjacent if they are 8-neighbors and have the same value in a binary image.
Simple Example: In a binary image, if a white pixel is diagonally touching another white pixel, they are 8-adjacent.

##### m-Adjacency (Mixed Adjacency):
m-Adjacent: Two pixels are m-adjacent if they are either 4-adjacent or diagonal neighbors (but not both), and they have the same or similar intensity values.
Improvement Over 8-Adjacency: This is considered better than 8-adjacency because sometimes 8-adjacency can be ambiguous. For example, in certain cases, diagonal connections might mistakenly connect two objects that should be separate.

##### Extension to Grayscale Images:
In grayscale images, adjacency is not just about having the exact same value. Pixels can be considered adjacent if their intensity values fall within a defined set or range, making the concept applicable to a broader range of images.

### 2.6 Digital Path
Digital path is a sequence of pixels where each pixel in the sequence is adjacent to the next pixel according to a defined adjacency criterion (4-adjacency, 8-adjacency, or m-adjacency). 
Digital paths are used to represent boundaries, shapes, and contours within an image.


A digital path P from pixel p to pixel q is a finite sequence of distinct pixels {p0, p1, ..., pn}, where p0 is p, pn is q, and pi is adjacent to pi+1 for all 0  i < n.

Usage: Digital paths are used to trace along the edges of objects, to define shapes, or to represent lines and curves within a digital image.



##### Closed Path
A digital path is considered a closed path if the starting pixel is the same as the ending pixel, meaning that it forms a loop.

Example: Imagine tracing around the edge of a circle on a digital image. If you start at one point on the circles edge and move along the edge until you return to the starting point, you have created a closed path.

Usage: Closed paths are especially important when dealing with shapes and objects that are enclosed, as they help in understanding the boundary and the area of the object.

##### Important of digital path
Boundary Representation: Digital paths are crucial for representing the boundaries of objects within an image, which is a foundational step in object recognition and image segmentation.

Shape Analysis: Closed paths allow for the analysis of the shape properties of an object, such as its perimeter, area, and circularity.

Contour Tracing: Digital paths are used in contour tracing algorithms, which are methods to extract the contour of an object from an image.
