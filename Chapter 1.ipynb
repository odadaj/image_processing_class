With the advent of powerful computing systems and software tools, this domain has seen tremendous growth and innovation.
```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

###  Loading images in Python using OpenCV

```python
# Load a sample image
img = cv2.imread("baboon.png")
```

###  Display various image types

```python
# Convert BGR image to RGB (since OpenCV loads images in BGR format)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# RGB Decomposition
r, g, b = cv2.split(img)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(r, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(b, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')
plt.show()
```

```python
# Convert to Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Convert to Binary Image (using a threshold of 128)
_, binary_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)
plt.imshow(binary_img, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
plt.show()
```

###  Matrix representation of images

```python
# Image as Matrix Representation
print("Shape of the Image Matrix:", gray_img.shape)
print("\nData type of the loaded image:", gray_img.dtype)
#print("\nFirst 10x10 values of the Red Channel in uint8 format:\n", binary_img[:10, :10, 0])
print("\nFirst 10x10 values of the Red Channel in uint8 format:\n", gray_img[:10, :10])
```

```python
# Convert to float32 for demonstration
img_float32 = np.float32(img) / 255.0
print("\nData type after conversion to float32:", img_float32.dtype)
print("First 5x5 values of the Red Channel in float32 format (normalized):\n", img_float32[:5, :5, 0])
```

```python
# Convert back to uint8
img_uint8_converted = np.uint8(img_float32 * 255)
print("\nData type after converting back to uint8:", img_uint8_converted.dtype)
print("First 5x5 values of the Red Channel after converting back to uint8:\n", img_uint8_converted[:5, :5, 0])
```

###  Typical Image Processing Operations

### 1. Binarization
Binarization is a process of converting a grayscale or color image into a binary image, where each pixel is assigned one of two possible values: black or white. It greatly simplifies the data of the image and can be crucial for tasks like Optical Character Recognition (OCR) or simplifying complex images.

```python
_, binary_image = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

In the code above, `cv2.threshold()` takes in the grayscale image and sets all pixels with a value greater than 127 to 255 (white) and all others to 0 (black).

### 2. Smoothing
Smoothing is used to reduce noise and small details in an image. There are various techniques for smoothing. By blurring the image, we can remove minor details or noise. This is often a preparatory step for other operations, ensuring they aren't affected by the minutiae. Here, we use the Gaussian Blur, which replaces each pixel's value with a weighted average of its neighbors.

```python
smoothed_image = cv2.GaussianBlur(img, (11,11), 0)
```

The `cv2.GaussianBlur()` function blurs the image, effectively reducing sharpness and noise.

### 3. Sharpening
Sharpening is the process of enhancing the edges and fine details in an image. This operation amplifies the contrast between pixels and can make images clearer for human viewers
This is usually achieved using a convolution operation with a sharpening kernel.

```python
kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpened_image = cv2.filter2D(img, -1, kernel)
```

The sharpening kernel amplifies the center pixel and reduces its neighbors, emphasizing differences.

### 4. Noise Removal
Real-world images often come with some form of noise, distortions caused by various factors from sensors to transmission errors. Noise removal filters can improve image clarity. Similarly, de-blurring attempts to correct images that are out of focus or affected by movement during capture.

Noise removal is essential for preparing images for further processing, especially if the images are captured under non-ideal conditions. The Median Blur is effective for removing "salt-and-pepper" noise.

```python
noise_removed_image = cv2.medianBlur(img, 5)
```

The `cv2.medianBlur()` function replaces each pixel's value with the median of its neighbors, which is effective at removing noise.

### 5. Edge Extraction
Before recognizing or analyzing the contents of an image, it's often useful to identify boundaries or edges within it. Edges in an image represent boundaries between different regions. Edge extraction highlights these boundaries, making them more evident and can be valuable for many applications like object detection. 

```python
edges = cv2.Canny(img, 100, 200)
```

The `cv2.Canny()` function detects edges by looking for regions in the image with a rapid change in intensity.

### 6. Segmentation
Segmentation divides an image into different parts or segments based on certain criteria, often to identify distinct objects or features in the image. It's a foundational step in object recognition and classification. One simple method is thresholding, which divides an image based on pixel intensity.

```python
blurred = cv2.GaussianBlur(gray_img, (7, 7), 0)
ret, segmented_image = cv2.threshold(blurred, 0, 255, cv2.THRESH_OTSU)
```

Otsu's method calculates an "optimal" threshold by maximizing the variance between two classes of pixels. 

```python
def display_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(2, 3, i+1)
        plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Display images
display_images(
    [binary_image, smoothed_image, sharpened_image, noise_removed_image, edges, segmented_image],
    ["Binarization", "Smoothing", "Sharpening", "Noise Removal", "Edge Extraction", "Segmentation"]
)
```
